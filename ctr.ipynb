```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import os

from google.colab import drive 
drive.mount('/content/gdrive')

dflt = '/content/gdrive/MyDrive/ctr'


year = '/2021년/'

xProc = '/코킹'
xPath = dflt+xProc+year
xFiles = os.listdir(xPath)

y1Proc = '/기동토크'
y1Path = dflt+y1Proc+year
y1Files = os.listdir(y1Path)

y2Proc = '/작동토크'
y2Path = dflt+y2Proc+year
y2Files = os.listdir(y2Path)

for x, y1, y2 in zip(xFiles, y1Files, y2Files):
  dfx = pd.read_csv(os.path.join(xPath, x), encoding = 'cp949', index_col=False)
  dfy1 = pd.read_csv(os.path.join(y1Path, y1), encoding = 'cp949', index_col=False)
  dfy2 = pd.read_csv(os.path.join(y2Path, y2), encoding = 'cp949', index_col=False)

  plt.figure()
  plt.subplot(4,1,1)
  plt.plot(dfx.코킹하중측정)
  plt.subplot(4,1,2)
  plt.plot(dfx.코킹거리측정)
  plt.subplot(4,1,3)
  plt.plot(dfy1.기동측정값)
  plt.subplot(4,1,4)
  plt.plot(dfy2.작동측정값) 

  # matplotlib.pyplot.pause()
  plt.waitforbuttonpress()

```
